```python
from matplotlib import pyplot as plt
```

# Student class

```python
class Student:

    def __init__(self, name, sub1=0, sub2=0, sub3=0, sub4=0, sub5=0, subject = ('OOPS', 'DM', 'MTCS', 'CSA', 'TC')):
        self.name = name
        self.subjects_name = subject
        self.sub1 = sub1
        self.sub2 = sub2
        self.sub3 = sub3
        self.sub4 = sub4
        self.sub5 = sub5

    def barPlot(self):
        '''
        Prints a bar plot for a particular student
        '''
        plt.bar(self.subjects_name,(self.sub1,self.sub2,self.sub3,self.sub4,self.sub5),width=0.3)
        plt.xlabel("Subjects")
        plt.ylabel("Marks")
        plt.title(f"Bar plot for {self.name}")
        plt.show()

    def histoPlot(self):
        fig = plt.figure(figsize = (10, 6))
        plt.hist((self.sub1,self.sub2,self.sub3,self.sub4,self.sub5))
        plt.xlabel("Marks")
        plt.ylabel("frequency")
        plt.title(f"Histogram plot for {self.name}")
        plt.show()

    def piePlot(self):
        explode = (0.07,0.03,0.15,0.08,0.1)
        plt.pie((self.sub1,self.sub2,self.sub3,self.sub4,self.sub5), labels = self.subjects_name, autopct ='%1.2f%%', explode = explode, shadow = True)
        plt.title(f"Pie plot for {self.name}")
        plt.show()
```

```python
def newStudent(students_objects_list):
    '''
    Inserts A new student in the students object list

    Attributes:
    ----------
        students_objects_list: the students object list to insert element in
    '''
    try:
        name = input("Enter Student Name")
        if not name:
            raise Exception("Empty string is not a valid name")
        marks = []        
        for i in ('OOPS', 'DM', 'MTCS', 'CSA', 'TC'):
            marks.append(float(input(f"Enter {i} marks for {name}: ")))
            if marks[-1]>100 or marks[-1]<0:
                raise Exception("Enter correct marks")
        
        stu = Student(name,marks[0],marks[1],marks[2],marks[3],marks[4])
        students_objects_list.append(stu)
    except Exception:
        print("Incorrect input")
```

```python
def stuBarPlot(students_objects_list):
    '''
    Prints a bar plot for a particular student
    
    Attributes:
    ----------
        students_objects_list: the students object list in which Students infromations are stored
    '''
    name = input("Enter Name of the student you want to see")
    if name not in (i.name for i in students_objects_list):
        print("No such student!!")
        return

    for i in students_objects_list:
        if i.name == name:
            i.barPlot()
        
```

```python
def stuPiePlot(students_objects_list):
    '''
    Prints a bar plot for a particular student
    
    Attributes:
    ----------
        students_objects_list: the students object list in which Students infromations are stored
    '''
    name = input("Enter Name of the student you want to see")
    if name not in (i.name for i in students_objects_list):
        print("No such student!!")
        return
        
    for i in students_objects_list:
        if i.name == name:
            i.piePlot()
```

```python
def stuHistoPlot(students_objects_list):
    '''
    Prints a bar plot for a particular student
    
    Attributes:
    ----------
        students_objects_list: the students object list in which Students infromations are stored
    '''
    name = input("Enter Name of the student you want to see")
    if name not in (i.name for i in students_objects_list):
        print("No such student!!")
        return
        
    for i in students_objects_list:
        if i.name == name:
            i.histoPlot()   
```

```python
def subBarPlot(students_objects_list):
    '''
    Prints a bar plot for a particular student
    
    Attributes:
    ----------
        students_objects_list: the students object list in which Students infromations are stored
    '''
    ch = int(input("Enter Subject choice(1-5):"))
    if ch>5 and ch<1:
        print("No such Subject!!")
        return

    students_names = list(i.name for i in students_objects_list)
    if ch == 1:
        plt.title(f"Bar plot for OOPS")
        subject_marks = list(i.sub1 for i in students_objects_list)
    elif ch == 2:
        plt.title(f"Bar plot for DM")
        subject_marks = list(i.sub2 for i in students_objects_list)
    elif ch == 3:
        plt.title(f"Bar plot for MTCS")
        subject_marks = list(i.sub3 for i in students_objects_list)
    elif ch == 4:
        plt.title(f"Bar plot for CSA")
        subject_marks = list(i.sub4 for i in students_objects_list)
    else:
        plt.title(f"Bar plot for TC")
        subject_marks = list(i.sub5 for i in students_objects_list)

    plt.bar(students_names,subject_marks,width=0.3)
    plt.xlabel("Subjects")
    plt.ylabel("Marks")
    plt.show()
```

```python
def subHistoPlot(students_objects_list):
    '''
    Prints a Histogram plot for a particular student
    
    Attributes:
    ----------
        students_objects_list: the students object list in which Students infromations are stored
    '''
    ch = int(input("Enter Subject choice(1-5):"))
    if ch>5 and ch<1:
        print("No such Subject!!")
        return

    # students_names = list(i.name for i in students_objects_list)
    if ch == 1:
        plt.title(f"Histogram plot for OOPS")
        subject_marks = list(i.sub1 for i in students_objects_list)
    elif ch == 2:
        plt.title(f"Histogram plot for DM")
        subject_marks = list(i.sub2 for i in students_objects_list)
    elif ch == 3:
        plt.title(f"Histogram plot for MTCS")
        subject_marks = list(i.sub3 for i in students_objects_list)
    elif ch == 4:
        plt.title(f"Histogram plot for CSA")
        subject_marks = list(i.sub4 for i in students_objects_list)
    else:
        plt.title(f"Histogram plot for TC")
        subject_marks = list(i.sub5 for i in students_objects_list)

    fig = plt.figure(figsize = (10, 6))
    plt.xlabel("Marks")
    plt.ylabel("frequency")
    plt.hist(subject_marks)
    plt.show()
```
